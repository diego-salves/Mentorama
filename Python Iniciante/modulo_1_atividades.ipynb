6. Escreva um programa para contar quantos números pares e ímpares existentes entre 1 e 10 bem como a soma deles.

a) usando a instrução while
b) usando a instrução for e as funções range e sum

```python
#a
conta_pares = 0
conta_impares = 0
x = 0
while x < 10:
    x += 1 #equivale a x = x + 1
    if x % 2 == 0:
        conta_pares += 1
    else:
        conta_impares += 1

print(f'Tem {conta_pares} números pares entre 1 e 10.')
print(f'Tem {conta_impares} números ímpares entre 1 e 10.')

#Aqui estou incluindo o 1 e 10 na contagem, se o resultado for 4 pares e 4 ímpares também está correto supondo-se que seriam os números ENTRE 1 e 10.
```

Dica: Iremos aprender sobre funções no próximo módulo, fique tranquilo. Contudo, você já
pode começar a praticar. A definição da função é a seguinte:

```python
#função

def bhaskara(a, b, c):
    delta = b ** 2 - 4 * a * c
if delta < 0:
    return None
else:
    raizes = []
    m1 = math.sqrt(delta)
    r1 =(-b + m1) / (2 * a)
    raizes.append(r1)
    r2 =(-b - m1) / (2 * a)
    raizes.append(r2)
    return raizes
```
